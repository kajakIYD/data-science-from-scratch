# Gradient Descent in linear regression
# Stochastic Gradient Descent in linear regression

```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from jupyterthemes import jtplot
# currently installed theme will be used to
# set plot style if no arguments provided
# jtplot.style()  # white plots :)

jtplot.style(theme='onedork')
```

```python
X = np.linspace(0, 10)
B = 3
b = np.zeros(len(X))
# display(X.shape, b.shape)
X = np.stack((X, b), axis=-1)
y = 2 * X[:, 0].T + B * X[:, 1]

m = len(X)

theta = np.random.random(2)
display(X.shape, y.shape, theta.shape)

delta_mses = []
thetas = []

eta = 0.01
for i in range(100):
    tmp = X.dot(theta) - y
#     display(tmp.shape)
    delta_mse = 2 / m * X.T.dot(tmp)
    delta_mses.append(delta_mse)
    thetas.append(theta)
    theta = theta - eta * delta_mse
    
plt.plot(delta_mses)
plt.show()

plt.plot(thetas)
plt.show()

```
